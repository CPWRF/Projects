```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel('ALSdata.xlsx')
df.tDateTime = pd.to_datetime(df.tDateTime, utc=True).dt.tz_convert('Asia/Taipei')
df = df[df.Item33St.isin([0,1])]
```

```python
plt.figure(figsize=(10,8))
sns.histplot(data=df[df.Item33>7000], x='Item33', hue='Item33St', bins=40, palette='RdYlGn')
plt.axvline(10000, color='blue', linestyle='--') #Test start in 07-22 10:51
plt.axvline(11000, color='blue', linestyle='--') # Move USL to 11000 from 10000 in 07-22 13:27
plt.axvline(12000, color='r', linestyle='--') # Move USL to 12000 from 11000 in 08-01 10:18
```
