```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel('baro.xlsx', index_col=8 ,parse_dates=True)
```

```python
df.info()
```

```python
df[df['value']==-999]['fixtureID2'].value_counts()
```

```python
df_fix6 = df[(df['fixtureID2']==6)]
df_fix4 = df[(df['fixtureID2']==4)]
```

```python
df_fix6.info()
```

```python
plt.figure(figsize=(30,10))
df_fix6.resample(rule='min').mean()['value'].plot()
plt.legend()
plt.savefig('fix4')
```

```python
plt.figure(figsize=(30,10))
df_fix4.resample(rule='min').mean()['value'].plot()
plt.savefig('fix6')
```
