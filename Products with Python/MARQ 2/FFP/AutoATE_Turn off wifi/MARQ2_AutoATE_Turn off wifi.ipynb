```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel('AutoATE_Turn off wifi.xlsx')
df.tDateTime = pd.to_datetime(df.tDateTime, utc=True).dt.tz_convert('Asia/Taipei')
df.sort_values('tDateTime',inplace=True)
df['TestProgramVersion'] = df.ExeInfo.apply(lambda info : info.split('[')[2][14:18])
```

```python
plt.figure(figsize=(12,8))
ax = sns.countplot(data=df, x='TestProgramVersion', hue='Item121St', palette='RdYlGn')
plt.legend(loc = 'upper left')
plt.title('AutoATE_Turn off wifi')

for p in ax.patches:
        ax.annotate('{:.0f}'.format(p.get_height()), (p.get_x()+0.1, p.get_height()+3))
```
