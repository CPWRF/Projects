```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
ct = pd.read_excel('data.xlsx',sheet_name='CT')
ht = pd.read_excel('data.xlsx',sheet_name='HT')
```

```python
ct.columns
```

```python
ht.columns
```

```python
plt.figure(figsize=(15,8))
plt.axhline(y=0, color='black', linestyle='--')
plt.axhline(y=0.5, color='b', linestyle='--')
plt.axhline(y=-0.5, color='b', linestyle='--')
sns.swarmplot(data=ct, x='StationID', y='GPS_L1 by ref', hue='Station', size=10)
plt.legend(loc="lower right")
plt.title('CT')
plt.savefig('CT_GPS_L1_byref' ,bbox_inches = "tight", facecolor='white', transparent=False, dpi=300)
```

```python
plt.figure(figsize=(15,8))
plt.axhline(y=0, color='black', linestyle='--')
plt.axhline(y=0.5, color='b', linestyle='--')
plt.axhline(y=-0.5, color='b', linestyle='--')
plt.axhline(y=2.5, color='r', linestyle='--')
plt.axhline(y=-2.5, color='r', linestyle='--')
sns.swarmplot(data=ht, x='StationID', y='GPS_L5 by ref', hue='Station', size=10)
plt.legend(loc="lower left")
plt.title('HT')
plt.savefig('HT_GPS_L5_byref' ,bbox_inches = "tight", facecolor='white', transparent=False, dpi=300)
```
