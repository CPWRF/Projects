```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel('data.xlsx')
```

```python
df.columns
```

```python
plt.figure(figsize=(15,8))
plt.axhline(y=-0.25, color='black',linestyle='--')
plt.axhline(y=0.1, color='b', linestyle='--')
plt.axhline(y=-0.6, color='b', linestyle='--')
plt.axhline(y=1.5, color='r', linestyle='--')
plt.axhline(y=-2, color='r', linestyle='--')
sns.swarmplot(data=df, x='stationID', y='C/No by ref.')
plt.savefig('CNo by ref' ,bbox_inches = "tight", facecolor='white', transparent=False, dpi=300)
```

```python
plt.figure(figsize=(10,8))
plt.axhline(y=-0.25, color='black',linestyle='--')
plt.axhline(y=0.1, color='b', linestyle='--')
plt.axhline(y=-0.6, color='b', linestyle='--')
plt.axhline(y=1.5, color='r', linestyle='--')
plt.axhline(y=-2, color='r', linestyle='--')
sns.swarmplot(data=df[df.stationID.isin([1,'1_New'])], x='stationID', y='C/No by ref.')
plt.savefig('CNo by ref_com1' ,bbox_inches = "tight", facecolor='white', transparent=False, dpi=300)
```
