```python
import pandas as pd
from plotnine import *
```

```python
raw = pd.read_excel('PR_PreMP_sonar_Tx_AVG.xlsx')
```

```python
(
    ggplot(raw.query("phase != 'PR'"))
    + aes(x='phase', y='Tx_avg')
    + geom_violin(style='left-right')
    + geom_point()
    + geom_line(aes(group='SerialNumber'))
    + geom_hline(yintercept = 666, color='red', linetype='--')
    # + geom_text(aes(label = 'Tx_avg'), size=10)
    + geom_text(aes(x=1,y=700),label='test', color='red')
    # + theme_minimal()
)
```
