```python
import pandas as pd
from plotnine import *
```

```python
raw = (pd.read_excel('PreMP_sonar_Tx_AVG_FindLSL.xlsx')
       .melt(id_vars='SerialNumber', value_vars=['per9degree','per180degree','per45degree']
         , value_name='Tx_AVG'
         , var_name='Rotation')
      #  .astype({'SerialNumber':'category'})
       )
```

```python
(
    ggplot(raw)
    + aes(x='Rotation', y='Tx_AVG')
    + geom_boxplot()
    + geom_point(aes(colour='SerialNumber'))
    + geom_line(aes(group='SerialNumber', colour='SerialNumber'), alpha=0.3)
    # + geom_text(aes(label='SerialNumber'))
    + geom_hline(yintercept=360, colour='red', linetype='--')
    + geom_text(x=0.7, y=380, label='360 as LSL', color='red', size=8)
    + geom_text(x=3.3, y=340, label='3460658835', size=8)
    + ggtitle('MK3 pre-MP first 69 units in 11/06~11/07')
)
```
