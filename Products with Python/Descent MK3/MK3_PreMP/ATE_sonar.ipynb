```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.style.use('bmh')
```

```python
raw = (pd.read_excel('ATE_sonar.xlsx')
        .sort_values('tDateTime')
        .drop_duplicates('SerialNumber')
        .dropna()
        )
```

```python
plt.figure(dpi=150)
plt.title('Sonar_Tx(XDCRP)')
sns.violinplot(data=raw, x='phase', y='Item151', inner='point');
plt.axhline(y=0.9, color='red', ls='--', lw=1)
plt.axhline(y=1.35, color='red', ls='--', lw=1)
```

```python
plt.figure(dpi=150)
plt.title('Sonar_Tx(XDCRN)')
sns.violinplot(data=raw, x='phase', y='Item152', inner='point');
plt.axhline(y=0.9, color='red', ls='--', lw=1)
plt.axhline(y=1.35, color='red', ls='--', lw=1)
```
