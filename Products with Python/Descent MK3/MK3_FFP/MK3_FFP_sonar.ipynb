```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.style.use('ggplot')
```

```python
raw = pd.read_excel('MK3_FFP_2.0_2.5_sonar.xlsx')
```

```python
def tweak_raw(raw):
    return(raw
    .sort_values('tDateTime')
    .drop_duplicates('SerialNumber', keep='last')
    )
```

```python
plt.figure(dpi=150)
plt.title('MK3_FFP_R/C thickness vs Sonar TX RMS Avg')
sns.histplot(data=tweak_raw(raw), x='Item37', hue='Thickness', kde=True)
```
