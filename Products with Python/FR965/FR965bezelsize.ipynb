```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import manufacturing as mn
```

```python
raw = pd.read_excel('FR965bezelsize.xlsx')
raw.head()
```

```python
plt.style.use('fivethirtyeight')
```

```python
g = sns.jointplot(data=raw, x='X', y='Y',xlim=(45.30, 45.57), ylim=(45.30, 45.57), height=10)
g.ax_joint.axhline(y=45.44, ls='--', color='r', lw=2)
g.ax_joint.axhline(y=45.56, ls='--', color='r', lw=2)
g.ax_joint.axvline(x=45.33, ls='--', color='r', lw=2)
g.ax_joint.axvline(x=45.45, ls='--', color='r', lw=2)
```

```python
mn.ppk_plot(data=raw['X'], lower_specification_limit=45.33, upper_specification_limit=45.45)
plt.show()
```
