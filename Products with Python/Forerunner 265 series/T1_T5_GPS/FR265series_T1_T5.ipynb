```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# FR265S

```python
fr265s = pd.read_excel('Raw data_PR build.xlsx', sheet_name='FR265S')
fr265 = pd.read_excel('Raw data_PR build.xlsx', sheet_name='FR265')
```

RT1 L1

```python
plt.style.use('fivethirtyeight')
```

```python
plt.figure(dpi=200);
sns.violinplot(data=fr265s[(fr265s.ItemNameType == 16920) & (fr265s.failitem.isin([0,189]))], x='ORG_ID', y='Item189(GPS_L1 by ref)');
plt.axhline(y=2, color='red',ls='--');
plt.axhline(y=-2, color='red',ls='--');
plt.title('FR265S_AutoRT1');
```
