```python
raw = pd.read_excel('45_ApproachS70.xlsx',sheet_name='Test Result')
```

```python
raw.sample(5)
```

```python
def filter(raw):
    return(raw
    .rename(columns = lambda name : name.replace(" ",'').replace('/','').replace('%',''))
    .astype({'ItemNameType':'category','ProcessType':'category','Item':'category','ItemName':'category'})
    # .assign(CountESN = lambda df : df.CountCountESN.str.split('/').str[1].astype('int16'))
    .query("~ProcessType.str.contains('Pack')")
    .query("~ProcessType.str.contains('Click')")
    .query("~ProcessType.str.contains('ProdScan')")
    .query("~ItemName.str.contains('ESN')")
    .query("~ItemName.str.contains('Check ProductionMap')")
    .query("~ItemName.str.contains('Fixture ID')")
    .query("Retest >= Retest.mean()")
    # .Retest.mean()
    )
filter(raw)
```
