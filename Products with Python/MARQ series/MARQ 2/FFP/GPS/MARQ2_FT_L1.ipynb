```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel('gpsdata.xlsx')
```

```python
df.tDateTime = pd.to_datetime(df.tDateTime, utc=True).dt.tz_convert('Asia/Taipei')
df['TestProgramVersion'] = df.exeinfo.apply(lambda info : info.split('[')[2][:-20])
```

FT_GPS_L1 by ref

```python
ft_L1_2 = df.loc[(df.ItemNameType==15688) & (df.Item20St.isin([0,1])),['SerialNumber','Item20St','Item20','Station','StationID','Version','TestProgramVersion','tDateTime']].sort_values(['SerialNumber','tDateTime'])
```

```python
NG_ESN = ft_L1_2.loc[ft_L1_2.Item20St == 0,'SerialNumber'].drop_duplicates()
```

```python
ft_L1_2_retry = ft_L1_2[ft_L1_2.SerialNumber.isin(NG_ESN)].sort_values(['SerialNumber','tDateTime'])
```

```python
ft_L1_2_retry = ft_L1_2_retry[(ft_L1_2_retry.tDateTime<'2022-07-28') & (ft_L1_2_retry.Station=='T1_5PR1AC02')]
ft_L1_2_retry.drop(index=[1249,1296,1297], inplace=True)
ft_L1_2_retry
```

```python
ft_L1_2_retry_d = ft_L1_2_retry.drop_duplicates(subset=['SerialNumber','Item20St'], keep='last')
ft_L1_2_retry_d
```

```python
ft_L1_2_retry_d.set_index('tDateTime').resample(rule='d')['Item20St'].agg(['sum','count'])
```

```python
plt.figure(figsize=(12,8))
sns.countplot(data=ft_L1_2_retry, x='StationID', hue='Item20St')
```
