```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
plt.style.use('fivethirtyeight')
```

```python
df = pd.read_excel('clickRatiodata.xlsx')
```

```python
df = df[df.ProductName.str.contains('MARQ|Magna')]
df.head()
```

```python
df_dup = df[df.duplicated(subset='SerialNumber',keep=False)].sort_values(['SerialNumber','tDateTime'])
df_dup.head()
```

```python
df_dup.drop_duplicates(subset='SerialNumber',keep='first')['ProductName'].value_counts()
```

```python
plt.figure(figsize=(10,8), dpi=200)
sns.kdeplot(data=df_dup.drop_duplicates(subset='SerialNumber',keep='first'),x='Item7', bw_method=0.3, color='red')
sns.kdeplot(data=df_dup.drop_duplicates(subset='SerialNumber',keep='last'),x='Item7',bw_method=0.3, color='blue')
plt.legend(labels=['First click', 'Last(~3times) click'])
plt.axvline(0.18, color='r', linestyle='--')
plt.xlabel('Button1_ClickRatio')
plt.title('ClickRatio in same batch of DUTs')
#plt.savefig('Click_Ratio' ,bbox_inches = "tight", facecolor='white', transparent=False, dpi=300)
```

```python
plt.figure(dpi=200)
sns.histplot(data=df_dup.drop_duplicates(subset='SerialNumber',keep='first'),x='Item7', binwidth=0.01, color='orange', alpha=0.3, kde=True)
sns.histplot(data=df_dup.drop_duplicates(subset='SerialNumber',keep='last'),x='Item7', binwidth=0.01, color='green', alpha=0.3, kde=True)
plt.legend(labels=['First click', 'Last(~3 times) click'])
plt.axvline(0.18, color='r', linestyle='--')
plt.xlabel('Button1_ClickRatio')
plt.title('ClickRatio in same 78 units')
plt.text(s='0.18', y=10, x=0.15, color='Red')
```
