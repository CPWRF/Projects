```python
import pandas as pd
import manufacturing as mn
```

```python
raw = pd.read_excel('EP42_clickratio.xlsx')
```

```python
def tweak_raw(raw):
    return(raw
    .query("ItemNameType == 12434")
    .sort_values('tDateTime', ascending=False)
    .drop_duplicates(subset='SerialNumber',keep='first')
    .melt(id_vars='SerialNumber', value_vars=['Item7','Item17','Item27','Item37','Item47'], value_name='ClickRatio')
    )
```

```python
tweak_raw(raw).to_excel('CR.xlsx')
```

```python
mn.ppk_plot(
tweak_raw(raw).ClickRatio,
lower_specification_limit= 0.17,
upper_specification_limit= 0.5,
show_dppm= True
)
```
