```python
from scipy.stats import hypergeom
import matplotlib.pyplot as plt
from numpy import vectorize
import pandas as pd
import plotly.express as px
```

Hypergeom

```python
#M : Total number of objects
#N : N drawn without replacement
#n : Total number of Type I objects
result = []
M = 1500 
for N in [12,24,48,96,192]: 
    for n in range(0, 200):
        rv = hypergeom(M, n, N)
        result.append([M, N, n, rv.pmf(0)])
df = pd.DataFrame(result, columns=['Total', 'N_drawn', 'NG_units', 'Accept_Prob'])
df.Accept_Prob = df.Accept_Prob * 100
```

AOQ

```python
def AOQ(Total, N_drawn, NG_units, Accept_Prob):
    return (Accept_Prob * (NG_units/Total) * (Total-N_drawn))/Total
df['AOQ'] = vectorize(AOQ)(df.Total, df.N_drawn, df.NG_units, df.Accept_Prob)
df['Escaped_unit'] = ((df.AOQ * M) / 100).round(2)
df.head()
```

plotly.express

```python
fig = px.line(df, x='NG_units', y='Accept_Prob', color='N_drawn', title=f"OC (operating characteristic) curve, population={M}")
fig.write_html("OCcurve_1500.html")
fig.show()
```

```python
fig2 = px.line(df, x='NG_units', y='Escaped_unit', color='N_drawn', title=f"AOQ (Average Outgoing Quality), population={M}")
fig2.write_html("AOQ_1500.html")
fig2.show()
```
