```python
from scipy.stats import hypergeom
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import plotly.express as px
```

```python
[M, n, N] = [200, 4, 6]
rv = hypergeom(M, n, N)
pmf= rv.pmf(0)
pmf
```

```python
#M : Total number of objects
#N : N drawn without replacement
#n : Total number of Type I objects

result = []

M = 200 

for N in [6,12,32, 80]: 
    for n in range(0, 100):
        rv = hypergeom(M, n, N)
        result.append([M, N, n, rv.pmf(0)])
```

```python
df = pd.DataFrame(result, columns=['Total', 'N_drawn', 'NG_units', 'Prob'])
df
```

```python
def AOQ(Total, N_drawn, NG_units, Prob):
    return (Prob * (NG_units/Total) * (Total-N_drawn))/Total
```

```python
df['AOQ'] = np.vectorize(AOQ)(df.Total, df.N_drawn, df.NG_units, df.Prob)
```

```python
df
```

```python
px.line(x=df['NG_units'], y=df['Prob'], color=df['N_drawn'], title='Accept_Prob')
```

```python
px.line(x=df['NG_units'], y=df['AOQ'], color=df['N_drawn'], title='AOQ')
```
