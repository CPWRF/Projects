```python
from scipy import stats
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
plt.style.use("fivethirtyeight")
```

```python
plt.hist(stats.norm.rvs(size=1000));
```

```python
plt.bar([k for k in range(4,10)],stats.binom.pmf(p=0.92, n=10, k=[k for k in range(4,10)]))
```

```python
poisson_pmf = stats.poisson.pmf(mu=4, k=[k for k in range(20)])
plt.bar([k for k in range(20)], poisson_pmf);
plt.xlim([0,20]);
```

```python
bernoulli_pmf =  stats.bernoulli.pmf(p=0.75, k=[k for k in range(2)])
plt.bar([0,1],bernoulli_pmf)
```

```python
stats.expon.pdf(0.000412*1500)
```

```python
expon_pdf =  stats.expon.pdf([0.000412*k for k in range(0,10000,2)])
plt.plot(expon_pdf);
```
