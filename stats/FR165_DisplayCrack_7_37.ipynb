```python
from scipy import stats as st
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
plt.style.use('bmh')
```

Binom model

```python
n=15
p=7/37
binom_model = st.binom(n=n, p=p)
```

```python
x = np.arange(0,10)
plt.figure(dpi=150)
sns.barplot(x=x, y=binom_model.pmf(x))
plt.title(f'n={n}')
```
