```python
from scipy import stats as st
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('ggplot')
```

```python
beta_dis = st.beta(a=1, b=30)
x = np.linspace(0,1,100)
plt.plot(x, beta_dis.pdf(x))
plt.xlim(0,0.3)
```

```python
beta_dis.cdf(0.1)
```

```python
plt.plot(x,[beta_dis.cdf(i) for i in x])
```
