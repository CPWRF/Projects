```python
from scipy import stats as st
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.style.use('bmh')
```

Binom model

```python
n = 20
p = 0.15
binom_model = st.binom(n=n, p=p)
```

```python
x = np.arange(0,10)
plt.figure(dpi=150)
plt.bar(x=x, height=binom_model.pmf(x))
plt.title(f'p = {round(p,2)}, n={n}')
```
