Beta

```python
a,b = 1, 29
x = np.linspace(0, 1.0, 10000)
dist_beta = beta(a,b)

plt.figure(dpi=150)
plt.plot(x, dist_beta.pdf(x), color='k');
plt.xlim([0,0.15])
plt.fill_between(x, beta.pdf(x, a, b), where=x<0.1)
plt.fill_between(x, beta.pdf(x, a, b), where=x<0.05)
plt.fill_between(x, beta.pdf(x, a, b), where=x<0.03)
plt.text(x=0.01, y=1, s=f'{100*dist_beta.cdf(0.03).round(2)}%')
plt.text(x=0.035, y=1, s=f'{100*dist_beta.cdf(0.05).round(2)}%')
plt.text(x=0.065, y=1, s=f'{100*dist_beta.cdf(0.1).round(2)}%')
plt.title(f'Beta distribution, a={a}, b={b}')
plt.xlabel('probability of success')
plt.ylabel('PDF')
```
