```python
plt.bar([7,8,9],binom_pmf);
```

```python
from scipy import stats
```

```python
stats.linregress(x=[1,4,5,2,9,4,10],y=[-3,18,25,4,53,18,60])
```

```python
stats.chisquare(f_obs=[60,80,65,70,80,100,140,120,90,90,60,65], f_exp=85)
```

```python
stats.t.pdf([1,2,3], 2)
```

```python
xs = np.linspace(-5,10,1000)
```

```python
plt.style.use("fivethirtyeight")
```

```python
plt.plot(xs, stats.t.pdf(xs, 1), label='df=1', ls='--', lw=1)
plt.plot(xs, stats.t.pdf(xs, 2), label='df=2', ls='--', lw=1)
plt.plot(xs, stats.t.pdf(xs, 3), label='df=3', ls='--', lw=1)
plt.plot(xs, stats.t.pdf(xs, 10), label='df=10', ls='--', lw=1)
plt.plot(xs, stats.t.pdf(xs, 30), label='df=30', ls='--', lw=1)
plt.legend()
```
