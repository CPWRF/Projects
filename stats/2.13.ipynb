```python
from scipy import stats as st
import matplotlib.pyplot as plt
import numpy as np
plt.style.use('fivethirtyeight')
```

Binom model

```python
n=210
p = 949/44554
binom_model = st.binom(n=n, p=p)
```

```python
x = np.arange(0,10)
plt.figure(dpi=150)
plt.bar(x=x, height=binom_model.pmf(x))
plt.title(f'p = {round(100*p,2)}%, n={n}')
plt.xlabel('k')
plt.ylabel('p')
plt.axhline(0.05, color='k', ls='--', lw=2)
```
