```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
project_list_file = 'ProjectListExport_2022-05-20-09-13-56-082.xls'
cft_list_file = 'CFTListExport_2022-05-20-09-10-10-246.xls'
prweb_file = 'New Production Status Summary 2022-05-24.xlsx'
```

```python
project = pd.read_excel(project_list_file, usecols=[0,1,3,4,5,6,9,10], parse_dates=[2,3,4,5], index_col=[0])
cft = pd.read_excel(cft_list_file, usecols=[0,44], index_col=[0])
```

```python
cft
```

```python
project = project[(project['Global/Asia'] == 'Global') & (project['PDP/Form107'] == 'PDP')]
project
```

```python
project_cft = pd.merge(left=project, right=cft, left_index=True, right_index=True)
project_cft.sort_index()
```

```python
consumer_dpqe_list = ['Lo, Aalto','Kuo, Chris','Wang, Cooper', 'Fu, Ernest','Fu, Ernest;\nLo, Aalto','Hsu, Jason(PQE)','Ni, Wendy']
```

```python
df_consumer_dpqe = project_cft[project_cft['Develop  Product QE (DPQE)'].isin(consumer_dpqe_list)].sort_index()
df_consumer_dpqe
```

```python
value_vars = ['Working Prototype Built and Tested','Final Factory Prototypes Built','PR Date','MP Date']

melted = df_consumer_dpqe.reset_index().melt(id_vars= ['Project ID','Project Name','Develop  Product QE (DPQE)'],value_vars=value_vars)
melted
```

```python
melted.rename(columns={'Develop  Product QE (DPQE)':'DPQE','variable':'Phase','value':'Date'},inplace=True)
```

```python
melted.Date = melted.Date.dt.strftime('%Y-%m')
```

```python
melted = melted[melted.Date > '2022-05']
```

```python
melted.dropna(inplace=True)
```

```python
# Weight socre, WP MP = 1, FP = 2, PR = 3
melted['weighted_score'] = melted['Phase'].apply(lambda phase : 1 if (phase == 'Working Prototype Built and Tested') | (phase == 'MP Date') else 2 if phase=='Final Factory Prototypes Built' else 3 if phase=='PR Date' else 0)
```

```python
prweb = pd.read_excel(prweb_file, header=1, usecols=[0,1,2,3])
```

```python
prweb
```

```python
multi_project = prweb.groupby(['PM Project Id','PM Project Name']).count()['Project Name'].sort_values(ascending=False).reset_index().rename(columns={'PM Project Id':'Project ID','Project Name':'Project_Qty_in_PRweb'})
```

```python
# Handle errors in PRweb
multi_project.loc[multi_project['PM Project Name']=='Forerunner 255 and 255 Music','Project_Qty_in_PRweb']=1
multi_project.loc[multi_project['PM Project Name']=='Descent Mk3 Series','Project_Qty_in_PRweb']=1
```

```python
multi_project[multi_project.Project_Qty_in_PRweb>1]
```

```python
final = pd.merge(how='left',left=melted, right=multi_project, on='Project ID')
```

```python
final.Project_Qty_in_PRweb = final.Project_Qty_in_PRweb.fillna(1)
```

```python
final['Weighted_Project_Qty_in_PRweb']=final['weighted_score'] * final['Project_Qty_in_PRweb']
```

```python
final.groupby(['DPQE','Project Name','Date']).sum()[['Project_Qty_in_PRweb','Weighted_Project_Qty_in_PRweb']].unstack().to_excel('(Weighted)Project_Qty_in_PRweb.xlsx')
```

```python
plt.figure(figsize=(20,15))
sns.heatmap(final.groupby(['DPQE','Date']).sum()['Project_Qty_in_PRweb'].unstack(),cmap="OrRd", annot=True,linewidths=0.5, linecolor='orange').set_title('Project_Qty_in_PRweb');
```

```python
plt.figure(figsize=(13,18))
sns.heatmap(final.groupby(['DPQE','Project Name','Date']).sum()['Project_Qty_in_PRweb'].unstack(),cmap="OrRd", annot=True,linewidths=0.5, linecolor='orange').set_title('Project_Qty_in_PRweb');
```

```python

for member in consumer_dpqe_list:
    plt.figure(figsize=(13,10))
    sns.heatmap(final.groupby(['DPQE','Project Name','Date']).sum()['Project_Qty_in_PRweb'].unstack(fill_value=0).loc[member],cmap="OrRd", annot=True).set(title=member+'_Project_Qty_in_PRweb');
```
