```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
project_list_file = 'ProjectListExport_2022-05-20-09-13-56-082.xls'
#cft_list_file = 'CFTListExport_2022-05-20-09-10-10-246.xls'
```

```python
project = pd.read_excel(project_list_file, usecols=[0,1,3,4,5,6,9,8,10], parse_dates=[2,3,4,5], index_col=[0])
#cft = pd.read_excel(cft_list_file, usecols=[0,44], index_col=[0])
```

```python
project = project[(project['Global/Asia'] == 'Global') & (project['PDP/Form107'] == 'PDP')]
project
```

```python
value_vars = ['Working Prototype Built and Tested','Final Factory Prototypes Built','PR Date','MP Date']

melted = project.reset_index().melt(id_vars= ['Project ID','Project Name','Category'],value_vars=value_vars)
melted
```

```python
melted = melted[(melted['variable'] != 'Working Prototype Built and Tested')]
```

```python
melted.dropna(inplace=True)
```

```python
melted['score'] = melted['variable'].apply(lambda phase: 2 if phase=='Final Factory Prototypes Built' else 3 if phase=='PR Date' else 1 if phase == 'MP Date' else np.nan)
```

```python
melted['date'] = melted['value'].dt.strftime('%Y-%m')
```

```python
after_june = melted[(melted.date >'2022-06-01') & (melted.date <'2022-12-31')]
```

```python
result = after_june.groupby(['Category','Project Name','date']).sum()['score'].unstack()
result.to_excel('after_june.xlsx')
```
