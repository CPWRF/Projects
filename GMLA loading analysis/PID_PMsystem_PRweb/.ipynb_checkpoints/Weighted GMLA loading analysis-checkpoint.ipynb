```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel('GMLA.xlsx', sheet_name='GMLAL Status Summary- 2022 New')
```

```python
df = df.iloc[:,:10]
```

```python
df.rename(columns={df.columns[6]: "Close meeting Target date", 
                   df.columns[7]: "Close meeting Actual date",
                   df.columns[8]: "GMLA start date",
                   df.columns[1]: "Project Name"}, inplace=True)
```

```python
df.drop((df.columns[3:5]), axis=1, inplace=True)
```

```python
df
```

```python
# df = df[['GMLA Phase','Close meeting Target date','DPQE']]
```

```python
df = df[df['Close meeting Target date'].notnull()]
```

```python
df = df[df['Close meeting Target date']>'2022-01-01']
```

```python
df
```

```python
df = df.set_index('Close meeting Target date')
```

```python
df['weighted'] = df['GMLA Phase'].apply(lambda gmla : 1 if gmla=='GMLA1' else 2 if gmla=='GMLA2' else 3)
```

```python
t1_dpqe = ['Cooper Wang', 'Ernest Fu','Aalto Lo', 'Ernest Fu/ Aalto Lo', 'Cooper Wang/ Wendy Ni']
aoem_dpqe = ['Simon Ko','Ray Hu','Cherry Tsai']
```

```python
df_aoem = df[df['DPQE'].isin(aoem_dpqe)]
df_consumer = df[(~df['DPQE'].isin(aoem_dpqe)) & (df['DPQE']!='Not assigned yet')]
```

```python
df_aoem = df_aoem.groupby('DPQE').resample(rule='m').sum()
```

```python
df_aoem
```

```python
df_aoem.unstack()['weighted'].transpose().plot.area(alpha=0.4,figsize=(15,10), grid=True,title='AOEM DPQE')
plt.xlabel('GMLA Target date')
plt.ylabel('GMLA weighted qty')
#plt.savefig('AOEM.jpeg', dpi=300)
```

```python
# Consumer DPQE
```

```python
df_consumer = df_consumer.groupby('DPQE').resample(rule='m').sum()
```

```python
df_consumer
```

```python
df_consumer.unstack()['weighted'].transpose().plot.area(alpha=0.4,figsize=(15,10), grid=True, title='Consumer DPQE')
plt.xlabel('GMLA Target date')
plt.ylabel('GMLA weighted qty')
# plt.savefig('consumer.jpeg', dpi=300)
```

```python
Ernest = df[df['DPQE']=='Ernest Fu']
```

```python
Ernest = Ernest.groupby('Project Name').resample('M').sum()
```
